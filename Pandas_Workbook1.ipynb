## Handling duplicates

```python
duplicate_df = movies_df.append(movies_df)
duplicate_df.shape
```

This adds a duplicate row with .append then checks it with .shape

```python
duplicate_df = duplicate_df.drop_duplicates()
duplicate_df.shape
```

This removes duplicates using .drop_duplicates

```python
duplicate_df.drop_duplicates(inplace=True)
```

It's a little verbose to keep assigning DataFrames to the same variable like in this example. For this reason, pandas has the inplace keyword argument on many of its methods. Using inplace=True will modify the DataFrame object in place
<br> Now temp_df will have the transformed data automatically.

```python
duplicate_df = movies_df.append(movies_df)

duplicate_df.drop_duplicates(inplace=True, keep=False)

duplicate_df.shape
```

The keep= argument has three options:
* First: Only keeps the first occurence
* Last: Only keeps the last occurence
* False: Deletes all occurences of duplicates
    * False is good if you want to selectively find duplicates

## How to work with missing values

There are two options in dealing with nulls:
1. Get rid of rows or columns with nulls
2. Replace nulls with non-null values, a technique known as imputation

Check which cells are null

```python
movies_df.isnull()
```

To calculate the number of nulls in each column

```python
movies_df.isnull().sum()
```

## Removing null values

```python
movies_df.dropna()
#Could specify (inplace=True)
```

Other than dropping rows, can also drop columns with null values by setting axis=1

```python
movies_df.dropna(axis=1)
#This would drop the revenue_millions and metascore columns (if the null values hadnt already been removed)
#It is 1 because in the tuple (rows,columns) columns are at index 1
```

## Imputation
Usually replace with the mean or median of that column

First, extract the column with nill values into its own variable

```python
revenue = movies_df['revenue_millions']
revenue.head()
```

```python
revenue_mean = revenue.mean()
revenue_mean
```

With the mean, fill nulls 

```python
revenue.fillna(revenue_mean, inplace=True)
```

By using **inplace=True** we have affected the original movies_df

To check

```python
movies_df.isnull().sum()
```

## Conditional selections

Take a column from the DataFrame and apply a Boolean condition to it

```python
condition = (movies_df['director'] == 'Ridley Scott')

condition.head()
```

To filter out all movies not directed by Ridley Scott

```python
movies_df[movies_df['director'] == 'Ridley Scott']
```

Conditional selections using numerical values by filtering the DataFrame by ratings

```python
movies_df[movies_df['rating'] >= 8.6].head(3)
```

We can make some richer conditionals by using logical operators | for "or" and & for "and".

```python
movies_df[(movies_df['director'] == 'Christopher Nolan') | (movies_df['director'] == 'Ridley Scott')].head()
```

Using **.isin()** to make this more consise

```python
movies_df[movies_df['director'].isin(['Christopher Nolan','Ridley Scott'])].head()
```

Let's say we want all movies that were released between 2005 and 2010, have a rating above 8.0, but made below the 25th percentile in revenue

```python
movies_df[
    ((movies_df['year'] >= 2005) & (movies_df['year'] <= 2010))
    & (movies_df['rating'] > 8.0)
    & (movies_df['revenue_millions'] < 17.442500)
]

#Lower quartile could have also been done 
#(movies_df['revenue_millions'] < movies_df['revenue_millions'].quantile(0.25))
```
